```python
    
def start_following():
    # to check whether we are already following or not we will try to find the follow btn, if it's there then you are not already following
    follow_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/section/main/div/header/section/div[1]/div[1]/div/div[2]/button/div/div')
    
    if follow_btn.get_attribute('innerHTML')=='Follow':
        follow_btn.click()
        time.sleep(1)
        print("You are following So Delhi now")
    else:
        print("You are already following")
        
# as after following we need to unfollow also so function for unfollow is
def unfollow():
    time.sleep(5)
    # we will check if follow is still written there on the button then you need to follow but if follow is there
    # that means you have already unfollowed else you need to unfollow
    follow_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/section/main/div/header/section/div[1]/div[1]/div/div[2]/button/div/div')
    
    if follow_btn.get_attribute('innerHTML')!='Follow':
        # you need to unfollow now and for that you need to find and click unfollow btn and click on it
        unfollow_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/section/main/div/header/section/div[1]/div[1]/div/div[2]/button')
        unfollow_btn.click()
        
        time.sleep(1)
        # now a popup window comes up to conferm that you want to unfollow so we click on unfollow to confirm
        driver.find_element(By.XPATH,'//button[contains(@class,"_a9-_")]').click()
        time.sleep(2)
        print("You have unfollowed So Delhi")
    else:
        print("You have already unfollowed")
        
#execute
if __name__ == "__main__":
    start_following()
    unfollow()
```

```python
    
def search_and_open():
    search_bar = driver.find_element(By.CLASS_NAME,'_aauy')
    search_bar.send_keys('dilsefoodie')
    time.sleep(2)
    #we are looking for the first option in search list so we will find first element with class '_abm4' and click on it
    driver.find_element(By.CLASS_NAME,'_abm4').click()
    time.sleep(3)
    # now we will open the first picture to like it
    first_pic = driver.find_element(By.CLASS_NAME,'_aagw')
    first_pic.click()
    #now we have open the first picture
    
def like_pictures():
    # we will start liking the pictures now
    count=0
    while count < 30:
        try:
            time.sleep(2)
            like_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div/div/div[2]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/button/div[2]')
            data = BeautifulSoup(like_btn.get_attribute('innerHTML'),'html.parser')
            if(data.find('svg')['aria-label'] == 'Like'):
                like_btn.click()
            time.sleep(2)

            # now we have to got next image 
            if count == 0:
                next_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div/div/div[1]/div/div/div/button/div')
            else:
                next_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div/div/div[1]/div/div/div[2]/button/div')
            next_btn.click()

            # we also have to keep the count of liked images
            count+=1
        except NoSuchElementException:
            print("Already liked the picture")
        
# after liking the pictures we also need to dislike the pictures
def unlike_pictures():
    # in order to dislike all the pictures
    # we will start liking the pictures now
    count=0
    while count < 30:
        try:
            time.sleep(2)
            like_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div/div/div[2]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/button/div')
            data = BeautifulSoup(like_btn.get_attribute('innerHTML'),'html.parser')
            if(data.find('svg')['aria-label'] == 'Unlike'):
                like_btn.click()
            time.sleep(2)

            # now we have to got next image 
            if count == 0:
                next_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div/div/div[1]/div/div/div/button/div')
            else:
                next_btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div/div/div[1]/div/div/div[2]/button/div')
            next_btn.click()

            # we also have to keep the count of liked images
            count+=1
        except NoSuchElementException:
            print("Already unliked the picture")
        
#execute 
if __name__ == "__main__":
    search_and_open()
    like_pictures()
    unlike_pictures()
    
```

```python
    
def open_followers_list(name):
    search_bar = driver.find_element(By.CLASS_NAME,'_aauy')
    search_bar.send_keys(name)
    time.sleep(2)
    #we are looking for the first option in search list so we will find first element with class '_abm4' and click on it
    required_name = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.CLASS_NAME,'_abm4')))
    required_name.click()
    time.sleep(3)
    # now we will locate where we can find followers list and click on it
    followers = WebDriverWait(
    driver, 20).until(EC.presence_of_element_located(
        (By.XPATH, "/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/section/main/div/header/section/ul/li[2]/a/div")))
    followers = driver.find_elements(By.CLASS_NAME,'_ac2a')[1]
    followers.click()
    time.sleep(4)
    #now we will have the followers list opened up
    
def return_to_homepage():
    driver.back()
    time.sleep(4)
    driver.back()
    time.sleep(2)
    #clear the search bar
    driver.find_element(By.CLASS_NAME,'_aauy').clear()
    #ckick the closs div so that search bar return to it's initial position
    try:
        cross_div = WebDriverWait(
            driver, 3).until(EC.presence_of_element_located(
                (By.XPATH, '//div[contains(@class,"_9-lv")]')))
        driver.find_element(By.XPATH,'//div[contains(@class,"_9-lv")]').click()
    except TimeoutException:
        print("search bar is clear")

def followers_of_sodelhi():
    # now we will check for the presence of popup window of followers
    pop_up_window = WebDriverWait(
    driver, 5).until(EC.presence_of_element_located(
        (By.XPATH, "/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[2]")))
    
    # now will scroll through the list and store all followers in a list
    followers_list = []
    previous_length = 0
    while True:
        pop_up_window = WebDriverWait(
        driver, 5).until(EC.presence_of_element_located(
            (By.XPATH, "/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[2]")))

        # now will scroll through the list and store all followers in a list
        followers_list = []
        previous_length = 0
        #data = BeautifulSoup(driver.page_source)
        while len(followers_list) < 500:
            #time.sleep(1)
            data = BeautifulSoup(driver.page_source)
            driver.execute_script(
                'arguments[0].scrollTop = arguments[0].scrollHeight;', 
              pop_up_window)
            time.sleep(3)
            followers_list=data.find_all(class_='_ab8w _ab94 _ab97 _ab9f _ab9k _ab9p _ab9- _aba8 _abcm')
        
    # here in am explicitly storing the names in a list so that i can display the names as list
    followers_500 = []
    for i in followers_list:
        followers_500.append(i.find(class_='_ab8w _ab94 _ab99 _ab9h _ab9m _ab9o _abcm').find('a').span.div.string)
        
    #now print the fllowers list
    print("Followers of SoDelhi :")
    print(followers_500)
    
    #as now we also need to find the followers for foodtalk india so we will go back to search page
    return_to_homepage()
    
def followers_of_foodtalkindia():
    # here in am explicitly storing the names in a list so that i can display the names as list
    pop_up_window = WebDriverWait(
    driver, 5).until(EC.presence_of_element_located(
        (By.XPATH, "/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[2]")))
    
    # now will scroll through the list and store all followers in a list
    followers_list = []
    previous_length = 0
    #data = BeautifulSoup(driver.page_source)
    while True:
        #time.sleep(1)
        data = BeautifulSoup(driver.page_source)
        driver.execute_script(
            'arguments[0].scrollTop = arguments[0].scrollHeight;', 
          pop_up_window)
        time.sleep(3)
        followers_list=data.find_all(class_='_ab8w _ab94 _ab97 _ab9f _ab9k _ab9p _ab9- _aba8 _abcm')
        # here in the followers list we can fech limited no of followers only because the followers list of foodtalkindia is 
        # private, so i have applied this if condition to break my loop when i have fetched all the data
        if len(followers_list) == previous_length or len(followers_list) > 500:
            break
        previous_length = len(followers_list)
    
    # here in am explicitly storing the names in a list so that i can display the names as list
    followers_500 = []
    for i in followers_list[:500]:
        for i in followers_list:
            followers_500.append(i.find(class_='_ab8w _ab94 _ab99 _ab9h _ab9m _ab9o _abcm').find('a').span.div.string)
            
    #now print the fllowers list
    print("Followers of foodtalkindia :")
    print(followers_500)
    return followers_500
    
# now we need to find the list of all the followers of “foodtalkindia” that you are following but those who don’t follow you
# for that first we will find the followers of foodtalkindia which we are following
def foodtalkindia_followers_followed_by_me(followers_list):
    followed_by_me = []
    for i in followers_list:
        if i.find('button').div.div.string == 'Following':
            followed_by_me.append(i.find(class_='_ab8w _ab94 _ab99 _ab9h _ab9m _ab9o _abcm').find('a').span.div.string)
        
    # now we will o back and fetch the list of our followers 
    return_to_homepage()
    
    # fetch the list of your followers
    #click on your profile
    driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/section/main/div[1]/section/div[3]/div[1]/div/div/div[2]/div/div/div/a').click()
    #click on followers
    time.sleep(2)
    driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/section/main/div/header/section/ul/li[2]/a/div').click()
    #list of my followers
    time.sleep(2)
    my_followers=driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[2]/div[1]/div/div')
    #we will store names in alist
    my_follower_list = []
    for i in my_followers:
        data=BeautifulSoup(i.get_attribute('innerHTML'))
        my_follower_list.append(data.find(class_='_ab8y _ab94 _ab97 _ab9f _ab9k _ab9p _abcm').string)
        
    # now to find required list of followers of foodtalkindia
    list_required = []
    for f in followed_by_me:
        bool = False
        for g in my_follower_list:
            if f == g:
                bool = True
        if bool == False:
            list_required.append(f)
            
    if len(list_required) == 0:
        print("no such followers")
    print(list_required)
        
        
#execute 
if __name__ == "__main__":
    #open followers list of sodelhi
    open_followers_list('sodelhi')
    followers_of_sodelhi()
    open_followers_list('foodtalkindia')
    # we need to take followers list of foodtalk india as a return for next part of our question
    followers_list = followers_of_foodtalkindia()
    # and we will pass that list
    foodtalkindia_followers_followed_by_me(followers_list)
```
